# Validation Dataset Viewer

Interactive viewer for browsing through the validation dataset images (Polish and Synth).

```python
from pathlib import Path

from datasets import load_from_disk

from dataset_viewer import DatasetViewer

# Project root
project_root = Path.cwd().parent
```

```python
# Load validation datasets
dataset_path = project_root / "data" / "datasets" / "validation" / "synth"

synth_dataset = load_from_disk(str(dataset_path))

print(f"Synth validation dataset loaded: {len(synth_dataset)} images")
```

```python
# Polish dataset viewer
synth_viewer = DatasetViewer(synth_dataset, title="Polish")
synth_viewer.show()
```
