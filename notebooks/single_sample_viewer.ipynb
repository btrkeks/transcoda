# Single Sample Viewer

Simple viewer for inspecting individual samples from the `antoniorv6/grandstaff-ekern` dataset.

**Usage:** Set `SAMPLE_INDEX` below and run all cells.

```python
# Configuration
SAMPLE_INDEX = 28279  # Index of the sample to view
SPLIT = "train"  # Dataset split: "train", "val", or "test"
```

```python
# Setup
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load the dataset
print(f"Loading dataset split: {SPLIT}...")
dataset = load_dataset("antoniorv6/grandstaff-ekern")
split_data = dataset[SPLIT]
print(f"Dataset loaded: {len(split_data)} samples in {SPLIT} split")
```

```python
# Display the sample
if SAMPLE_INDEX < 0 or SAMPLE_INDEX >= len(split_data):
    print(f"ERROR: Index {SAMPLE_INDEX} out of range [0, {len(split_data) - 1}]")
else:
    example = split_data[SAMPLE_INDEX]

    # Display image
    img = example["image"]
    transcription = example["transcription"]

    fig, ax = plt.subplots(figsize=(15, 10))
    ax.imshow(img)
    ax.axis("off")
    ax.set_title(f"Sample {SAMPLE_INDEX} from {SPLIT} split | Image size: {img.size}", fontsize=14)
    plt.tight_layout()
    plt.show()

    # Display transcription
    print("\n" + "=" * 80)
    print("TRANSCRIPTION:")
    print("=" * 80)
    print(transcription)
    print("=" * 80)
    print(f"\nTranscription length: {len(transcription)} characters")
    print(f"Number of lines: {len(transcription.splitlines())}")
```
