```python
import sys
from pathlib import Path

# Add project root to path
project_root = Path.cwd().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

import ipywidgets as widgets
import matplotlib.pyplot as plt
from ipywidgets import interact

from src.synthetic.generator import ScoreRenderOptions, VerovioGenerator

# Set up matplotlib for better display
plt.rcParams["figure.figsize"] = (15, 10)
plt.rcParams["figure.dpi"] = 100

print("Setup complete!")
```

```python
# Initialize the generator
# Change these parameters if needed
SOURCE_DATASET = "antoniorv6/grandstaff-ekern"
SPLIT = "train"

print("Initializing generator with:")
print(f"  Dataset: {SOURCE_DATASET}")
print(f"  Split: {SPLIT}")
print("\nThis may take a moment to load the dataset...")
generator = VerovioGenerator(sources=SOURCE_DATASET, split=SPLIT)
print("\n✓ Generator initialized successfully!")
```

```python
def generate_and_display(
    num_systems=3,
    reduce_ratio=0.5,
    trim_bottom_white=False,
    include_title=True,
    include_author=True,
    include_time_signature=True,
    show_transcription=False,
):
    """Generate and display an image with the given parameters."""
    options = ScoreRenderOptions(
        num_systems=num_systems,
        trim_bottom_white=trim_bottom_white,
        include_title=include_title,
        include_author=include_author,
        include_time_signature=include_time_signature,
        texturize_image=True,
        reduce_ratio=reduce_ratio,
    )

    image, transcription = generator.generate_full_page_score(options=options)

    plt.figure(figsize=(12, 8))
    plt.imshow(image)
    plt.axis("off")
    plt.title(f"Generated Score ({image.shape[1]}x{image.shape[0]}px, {len(transcription)} tokens)")
    plt.tight_layout()
    plt.show()

    if show_transcription:
        print(transcription)


# Create interactive widget
interact(
    generate_and_display,
    num_systems=widgets.IntSlider(min=1, max=7, step=1, value=3, description="Systems:"),
    reduce_ratio=widgets.FloatSlider(
        min=0.2, max=1.0, step=0.1, value=0.5, description="Reduce Ratio:"
    ),
    trim_bottom_white=widgets.Checkbox(value=False, description="Trim Bottom White"),
    include_title=widgets.Checkbox(value=True, description="Include Title"),
    include_author=widgets.Checkbox(value=True, description="Include Author"),
    include_time_signature=widgets.Checkbox(value=True, description="Include Time Sig"),
    show_transcription=widgets.Checkbox(value=False, description="Show Transcription"),
);
```
